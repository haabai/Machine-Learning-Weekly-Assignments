This notebook uses the provided `honey.csv` dataset to explore the health of the U.S. bee population across states and years. 


```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
honey = pd.read_csv('data\data/honey.csv')
honey.head()
```

```python
hn = honey[honey['state']=='AZ']
print(len(hn))
hn
```

```python
sns.relplot(x="prodvalue", y="state", data=honey, height=15, aspect=.5);
```

```python
sns.relplot(x="prodvalue", y="state", hue="year", palette="YlOrBr", data=honey, height=15, aspect=.5);
```

```python
sns.relplot(x="priceperlb", y="state", size="year", data=honey, height=15, aspect=.5);
```

```python
sns.relplot(x="year", y="prodvalue", kind="line", data=hn);
```

```python
sns.relplot(
    data=honey, 
    x="numcol", y="priceperlb",
    col="year", 
    col_wrap=2,
    kind="line"
)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['numcol'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Honey Production Year over Year');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['yieldpercol'], ax=ax2, color="r", 
                         label ='Yield per colony', legend=False) 
sns.despine(right=False)
plt.ylabel('colony yield')
ax.figure.legend();
```

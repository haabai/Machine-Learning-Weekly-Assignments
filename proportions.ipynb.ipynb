## Sample Dataset


```python
import pandas as pd

# Sample sales dataset
sales_data = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Groceries', 'Books', 'Toys'],
    'Sales': [10000, 7500, 16000, 5000, 2000]
})

print(sales_data)
```

```python
sales_data.to_excel('sales_data.xlsx', index=False)
```

```python
# Select a data range from the sales_data DataFrame (e.g., rows 1 to 3)
selected_range = sales_data.iloc[1:4]
print(selected_range)
```

```python
import matplotlib.pyplot as plt

# Create a 2-D pie chart for sales by category
plt.figure(figsize=(4, 4))
plt.pie(sales_data['Sales'], labels=sales_data['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Category')
plt.show()
```

##Create a Donut Chart using the same dataset

```python
# Create a donut chart for sales by category
plt.figure(figsize=(4, 4))
wedges, texts, autotexts = plt.pie(
    sales_data['Sales'],
    labels=sales_data['Category'],
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops=dict(width=0.4)
)
plt.title('Sales Distribution by Category (Donut Chart)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle

# Dataset
sales_data = pd.DataFrame({ 
    'Category': ['Electronics', 'Clothing', 'Groceries', 'Books', 'Toys'],
    'Sales': [10000, 7500, 16000, 5000, 2000]
})

# Normalize sales to percentages of 100 squares
total_sales = sales_data['Sales'].sum()
percentages = (sales_data['Sales'] / total_sales * 100).round().astype(int)

# Prepare data as dict for pywaffle
data_dict = dict(zip(sales_data['Category'], percentages))

# Plot waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    values=data_dict,
    colors=["#4c78a8", "#f58518", "#54a24b", "#e45756", "#b279a2"],
    title={'label': 'Sales Distribution by Category', 'loc': 'center'},
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    figsize=(10, 6)
)

plt.show()
```

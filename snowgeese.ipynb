```python
import pandas as pd
import matplotlib.pyplot as plt

birds = pd.DataFrame({
    'Name': ['Snow Goose'],
    'MaxWingspan': [165]  # values in centimeters
})

# Create bar chart
plt.figure(figsize=(6, 4))
plt.bar(birds['Name'], birds['MaxWingspan'], color='blue')

# Label and styling
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='both', which='both', labelbottom=True, bottom=True)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: highlight data point
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, 'bo')  # plot dot at the top of bar

plt.show()
```
